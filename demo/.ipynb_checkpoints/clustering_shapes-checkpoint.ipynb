# CLUSTERING SHAPES

```python
import numpy as np
import cv2
import os
import shapecontext
import morphology_utils
import multiprocessing
from joblib import Parallel, delayed
from sklearn.cluster import MiniBatchKMeans
```

Reading the images

```python
IM_DIR = "../data/shoe_images/"

images = []
for img in os.listdir(IM_DIR):
    if 'jpg' in img:
        path = os.path.join(IM_DIR, img)
        images.append(cv2.imread(path))

print len(images)
print images[0].shape
```

Compute the shape context bin histograms (in parallel)

```python
sc = shapecontext.ShapeContext()
def sc_array(img):
    sp = morphology_utils.shape_points(img, 30)
    bh = sc.compute(sp)
    return bh

num_cores = multiprocessing.cpu_count()
sc_arrays_list  = Parallel(n_jobs=num_cores)(delayed(sc_array)(img) for img in images)
sc_arrays = np.array(sc_arrays_list)
print sc_arrays.shape
```

5000 images and histograms with 8640 integers. Time to cluster them. 

For this particular excercise here we are using 5000 images. Back in the days, where I was developing this solution in the "real world" we had over 60000 active shoes in our website and well over 300000 shoe-images stored. Therefore, for speed reasons, we decided to use [MiniBatchKMeans ](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.MiniBatchKMeans.html). 

As we all know deciding the number of clusters when using kmeans is a tricky business. There are a few helping methods, such as the elblow curve or BIC (see [here](https://en.wikipedia.org/wiki/Determining_the_number_of_clusters_in_a_data_set)). We will use them here just to illustrate the use, but before, let's stop and think for a second.

Here we are dealing with 8640 integers vector, so we are way beyond the curse of dimensionality and is unlikely that these methods provide a definitive answer. In general, is unlikely that the shapes cluster very well. However, we do not need perfect clusters. In reality, we just need that a few clusters capture the side-views of the shoes. Then, we can concentrate in these clusters and ignore the others. In addition, for this particular example, we are dealing with 5000 shoe images (less observations than dimmensions!) which, added to the high dimensioanlity, are not enough shoe-images for patterns to clearly emeger using these techniques. 

Furthermore, the elblow curve or the BIC are statistical quantities and, although useful, one should not decide the number of clusters based on them. They are a good starting point, but the number of clusters should be decided based on the performace of your overall algorithm (if there are more elements other than kmeans). 

For example, in the first instance, this algorithm was used to show shoes in our site in a consistent way. Ultimately it was part of a content-based recommendation algorithm. Therefore, the number of clusters should be decided based on: $i)$ the increase of clicks/buys on the site and, of course, $i)$ a lot of exploration. Although tedious, optimising based on a final-task metric is fundamental.

Having said all of the above let's plot the elbow curve or the BIC score. 
You can find a version of the code below in stackoverflow, for example, [here](https://stats.stackexchange.com/questions/90769/using-bic-to-estimate-the-number-of-k-in-kmeans).

```python
from scipy.spatial.distance import cdist,pdist

def avg_within_ss(X, k):

    model = MiniBatchKMeans(init='k-means++', n_clusters=k, batch_size=50,
                          n_init=3, max_no_improvement=10, verbose=0)
    model.fit(X)

    centroids = model.cluster_centers_
    dist_c = cdist(X, centroids, 'euclidean')
    dist   = np.min(dist_c, axis=1)
    avgwithinss = sum(dist**2)/X.shape[0]

    return avgwithinss

def bic(X, k):

    model = MiniBatchKMeans(init='k-means++', n_clusters=k, batch_size=50,
                          n_init=3, max_no_improvement=10, verbose=0)
    model.fit(X)

    centers = model.cluster_centers_
    centers = np.expand_dims(centers, axis=1)
    labels  = model.labels_
    N_C = np.bincount(labels)
    R, M = X.shape

    wcss = sum([sum(cdist(X[np.where(labels == c)], centers[c], 'euclidean')**2) for c in range(k)])
    var = (1.0/(R-k)/M) * wcss
    const_term = 0.5 * k * np.log(R) * (M+1)

    BIC = np.sum([ ( Rn * np.log(Rn) ) -
                   ( Rn * np.log(R) ) -
                   ( ((Rn * M) / 2) * np.log(2*np.pi*var) )  -
                   ( (Rn - 1) * M/ 2 )
                   for Rn in N_C]) - const_term

    return BIC
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pylab
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

avgWithinSS = [avg_within_ss(sc_arrays, k) for k in range(5,40)]

# elbow curve
elbow_curve = plt.figure()
ax = elbow_curve.add_subplot(1,1,1)
ax.plot(range(5,40), avgWithinSS, 'b*-')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within cluster sum of squares')
plt.title('Elbow Curve for MiniBatchKMeans')
```

```python
bics = [bic(sc_arrays, k) for k in range(5,40)]

# bic score
bic_curve = plt.figure()
ax = bic_curve.add_subplot(1,1,1)
ax.plot(range(5,40), bics, 'b*-')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('BIC Score')
plt.title('BIC Score for MiniBatchKMeans')
```

As anticipated, not much information in the plots as to how many clusters should we chose. In fact, one might think that there might not be cluster structure in the data. However, I will emphasize again, we do not need perfect clusters, we just need that a few clusters capture the side view. For that, the data is good enough (well, maybe not 5000 shoe-images, but over 300000 certainly was). 

As this stage, I can tell you, using hundreds of thousands shoes and after a lot of exploration, we found that the "optimal" number of clusters was between 15 and 20. So, let's pick 15 and have a look to some clusters. 

```python
shoe_clusters = MiniBatchKMeans(init='k-means++', n_clusters=15, batch_size=50,
                                n_init=3, max_no_improvement=10, verbose=0)
shoe_clusters.fit(sc_arrays)
shoe_clusters_labels = shoe_clusters.labels_
```

Let's plot a few shoes for some clusters

```python
from random import sample
def plot_cluster(image_list, labels, c_id, n_images=20, cols=4):

    idx = np.where(labels == c_id)[0]
    idx = sample(idx, n_images)
    imgs = [image_list[i] for i in idx]

    nrow = len(idx) / cols
    ncol = cols

    if ((ncol * nrow) != len(idx)): nrow = nrow + 1

    plt.figure()
    for i,img in enumerate(imgs):
        plt.subplot(nrow,ncol,i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.show() 
```

```python
plot_cluster(images, shoe_clusters_labels, 2)
```

These cluster certainly seems to capture the side view and predominantly, trainers. Some flaws are clearly found within the cluster.

On the other hand one can see that some white-background shoe-images $--$ these are misclassifications $--$ and shoe-images with a grey, slightly-structured background. These are in this cluster because throughout the entire process, we did not perform a thorough background removal. In a last notebook I will show how this problem can be solved.

Finally, one may think that given the shape-point repetition I mentioned in the shape_context_algo notebook chances are that some of the dimensions/features are linearly related, and PCA can be a good option to reduce dimensionality, so why not try it?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
sc_arrays_sc = scaler.fit_transform(sc_arrays)

pc = PCA()
sc_arrays_pca = pc.fit_transform(sc_arrays_sc)
```

# REMOVE THE BACKGROUND

As we saw in the *clustering_shapes* notebook, some shoes have a slightly structured background. Removing this background was not part of our initial process for two reasons: $i)$ there might be reasons why one would like to automatically find the retailers that send images with a non-white background and $ii)$ removing the background is a computanionally expensive process. Therefore, we might prefer to remove the background after clustering the shapes of the entire dataset, and only when neccesary (we know that shoes with background will *all* belong to one of the clusters). Then we use the clustering model to find the cluster than the cleaned images belong to. 

Nonetheless, let's see how we can remove the background. First let's plot on of the shoes with a non-white background:

```python
import cv2
import pylab as pyl
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

def plot_image(image):
    if len(image.shape) == 2:
        pyl.imshow(image,cmap='gray')
    else:
        pyl.imshow(image)
    pyl.show()

shoe_image = cv2.imread("../data/shoe_images/42375516.jpg")
plot_image(shoe_image)
```

First let's get the edges using the [canny](http://docs.opencv.org/trunk/da/d22/tutorial_py_canny.html) edge detector in opencv

```python
import cv2

gray= cv2.cvtColor(shoe_image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,30,200)
plot_image(edges)
```

Now let's blur the image and find a threshold.

```python
blur = cv2.GaussianBlur(edges,(3,3),2)
_,thresh = cv2.threshold(blur,10,255,cv2.THRESH_BINARY)
adp_thresh = thresh.copy()
h,w = adp_thresh.shape[:2]

plot_image(adp_thresh)
```

Now we are goign to apply a mask based on a [flood fill](https://en.wikipedia.org/wiki/Flood_fill) algorithm, which is conveniently implemented in opencv as [cv2.floodFill()](http://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html#floodfill). 

There are a couple of quick things to mention before we move on. As one can read in the documentation: *"operation mask should be a single-channel 8-bit image, 2 pixels wider and 2 pixels taller than image"*

Also, cv2 reads coordinates in *"plot system-of-reference"*. This is, x runs horizontally to the right and y
vertically down the image. However numpy interprets the image inverted. This is, x runs vertically down the image while y runs horizontally to the right. So just be aware.

Finally, we set filled pixels to 125 (you can pick another convenient value). Let's go!

```python
mask = np.zeros((h+2,w+2),np.uint8)
if np.mean(thresh[:5,:5]) == 0:                   #upper left box
    cv2.floodFill(adp_thresh,mask,(0,0),(125))    #upper left point entry

if np.mean(thresh[h-6:h-1,:5]) == 0:              #lower left box
    cv2.floodFill(adp_thresh,mask,(0,h-1),(125))  #lower left point entry

if np.mean(thresh[h-6:h-1,w-6:w-1]) == 0:         # lower right box
    cv2.floodFill(adp_thresh,mask,(w-1,h-1),(125))# lower right point entry

if  np.mean(thresh[:5,w-6:w-1]) == 0:             # upper right box
    cv2.floodFill(adp_thresh,mask,(w-1,0),(125))  # upper right point entry
mask  = 255 - cv2.inRange(adp_thresh,125,125) 
plot_image(mask)
```

This is an easy case because the background was quite uniform, and we might decide to stop here. In some other cases, we might need to go deeper and apply an algorithm for foreground extraction, such as [GrabCut](http://docs.opencv.org/3.1.0/d8/d83/tutorial_py_grabcut.html). Let's see the result.

```python
# the mask to be passed to grabCut is encoded as follows (see documentation):
# 0 - absolutely backgroung
# 1 - absolutely foreground
# 2 - probably background
# 3 - probably foreground
g_mask = np.where((mask==255),3,2).astype('uint8')

bgdModel = np.zeros((1,65),np.float64) # default (see documentation) 
fgdModel = np.zeros((1,65),np.float64) # default
cv2.grabCut(shoe_image,g_mask,None,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_MASK)
g_mask = np.where((g_mask==2)|(g_mask==0),0,1).astype('uint8') # if 2 or 0 -> 0 (background)

plot_image(g_mask)
```

In fact, we have a slightly better result, altough we will be missing a small region in the front of the show. In this case, it does not matter, as it does not affect the overall shape.

So, we have the mask, let's apply it to the original image and check the result

```python
result = cv2.bitwise_and(shoe_image, shoe_image, mask = g_mask)

# back to 255 scale
result[result == 0] = 255

# let's see
plot_image(result)
```

And there we go!, the background is removed and the image is ready to go through the shape context algorithm!
